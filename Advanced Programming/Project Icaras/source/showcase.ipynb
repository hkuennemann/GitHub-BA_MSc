**Fifth**, we delve into national flight patterns by plotting routes from airports within a chosen country. This method categorizes flights into short-haul and long-haul based on distance, with an emphasis on potential environmental benefits of substituting short-haul flights with train travel. It enables a clear visualization of a country's flight network. Through this approach, we offer a unique perspective on the internal connectivity of countries, providing a visual and quantitative analysis of their aviation footprint and opportunities for emission reduction. While there are several definitions of short-haul flights, we have set it at 1000km.

Carbon Footprint Driving vs Flying Calculator (Gas vs Electric Car Footprint) - 8 BillionTrees.com

https://8billiontrees.com/carbon-offsets-credits/carbon-ecological-footprint-calculators/carbon-footprint-driving-vs-flying/

According to this link, the CO2 Emissions Per 100 km (in kg) is 25kg of CO2e per 100 km for planes, while trains only have CO2 Emissions Per 100 km (in kg) of 3 kg of CO2e per 100 km leading to a ratio of 3/25. Turkey, being a large country, yet having a very poor long-distance train system has a lot of potential to decrease the CO2 Emissions with building train tracks for short-haul routes. For that, we look at to how much percent, we can decrease the CO2 Emissions, if we were to replace all short-haul flights with train routes.

Note: For assessing CO2 emissions, it's necessary to count all flights individually, as emissions accumulate per flight regardless of repeated routes, thereby providing a more accurate environmental impact analysis. Thus, we consider all flights (consistently) in our analysis, as discussed in the lecture.

```python
flight.plot_country_flights("Turkey", internal=True)
```

When plotting the internal flights, we see that there are 250 short-haul flights in the dataset, compared to 56 long-haul flights. Due to this and also the total distance, we could reduce CO2 emissions to 37.15% of the current emissions, if we were to replace short-haul flights with rail services.

```python
flight.plot_country_flights("Turkey", internal=False)
```

Istanbul being an epicenter for travel it is obvious that there are many more long-haul flights, if we plot all flights starting in Turkey. Thus, the overall CO2 emissions reduce to "only" 91.07%, when looking at the bigger picture.
